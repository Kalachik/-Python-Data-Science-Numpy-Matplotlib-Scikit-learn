# Задание 2. Библиотека Matplotlib

```python
t = np.linspace(0,10,51) 
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t,f, color="green")

label_font = {"fontsize":11,
             "fontweight":"bold",
              'family':'arial',
            }
plt.title('График f(t)', fontweight ='bold', family ='arial' , loc = 'left')
plt.xlabel('Значения t', fontdict = label_font)
plt.ylabel('Значения f', fontdict = label_font)

plt.axis([0.5,9.5,-2.5,2.5])

plt.show()
```

# Задание 3. Библиотека Matplotlib

```python
x = np.linspace(-3,3,51)
x
```

```python
y1 = x**2
y2 = x*2+0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
y4
```

```python
fig,ax = plt.subplots (nrows =2 , ncols =2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title("График y2")
ax3.plot(x, y3)
ax3.set_title("График y3")
ax4.plot(x, y4)
ax4.set_title("График y4")

plt.show()
```

```python
cl_fn=df['Class'].value_counts()
cl_fn
```

```python
x=cl_fn.index
y=cl_fn[x]

fig,ax = plt.subplots (nrows =1 , ncols =2)
ax1, ax2 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.bar(x,y)
ax1.set_title("Столбчатая диаграмма")

ax2.bar(x,y, log =True)
ax2.set_title("Столбчатая диаграмма LOG")

plt.show()
y
```
